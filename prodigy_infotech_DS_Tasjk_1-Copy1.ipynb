```python
# filter data for total population
total_population_data=data[data['Indicator Code']=='SP.POP.TOTL']
total_population_data
```

```python
#sorting data
sort_population_data = total_population_data.sort_values(by='2022', ascending=False)
sort_population_data
```

# Top 5 countries with highest population in  2022


```python
top_5_countries = sort_population_data.head(5)
print("Top 5 countries with highest population is: ")
print(top_5_countries)
```

# Top 5 countries with highest population in 2022 and 2000


```python
# creat a bar plot
plt.figure(figsize=(20,15))
plt.rcParams['axes.facecolor']=="Skyblue"

# Bar plot for 2022
plt.subplot(2, 2, 1)
plot=sns.barplot(x='2022',y="Country Name",data=top_5_countries)
plt.xlabel('Total Population')
plt.ylabel('Country')
plt.title('Top 5 Countries by Population in 2022')

#bar plot for 2000
plt.subplot(2, 2, 2)
plot=sns.barplot(x='2000',y="Country Name",data=top_5_countries)
plt.xlabel('Total Population')
plt.ylabel('Country')
plt.title('Top 5 Countries by Population in 2022')
```
